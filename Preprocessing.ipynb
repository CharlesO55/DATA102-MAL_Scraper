```python
FOLDER = 'Detail'

FILENAMES = [f'{FOLDER}/{file}' for file in os.listdir('Detail')[1:]]
FILENAMES
```

```python
print(FILENAMES[1])
df = pd.read_csv(FILENAMES[1], index_col=0)
```

```python
# Drop useless/redundant columns
df = df.drop(columns=['German', 'Licensors', 'Spanish', 'Source', 'Demographic', 'Synonyms','Demographics', 'French', 'Duration', 'Members', 'Scorers', 'Related_Entries', 'Title'])
```

```python
# Merge the redundant columns while handling NaN values
df['Genre_Theme'] = df[['Genres', 'Genre', 'Theme', 'Themes']].apply(lambda x: ', '.join(x.dropna()), axis=1)

# Drop the original columns
df = df.drop(columns=['Genres', 'Genre', 'Theme', 'Themes'])
```

```python
df.columns
```

```python
df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce').fillna(0).astype(int)

df['Favorites'] = df['Favorites'].str.replace('[#,]', '', regex=True).astype(int)

df['Score'] = df['Score'].str.replace('[', '0').str.replace(']', '0').astype(float)

df['Favorites'] = df['Favorites'].astype(int)

df['Score'] = df['Score'].astype(float)

df['Ranked'] = df['Ranked'].str.replace('[#,]', '', regex=True).astype(float)
df['Popularity'] = df['Popularity'].str.replace('[#,]', '', regex=True).astype(float)
```
